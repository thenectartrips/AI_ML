```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
print("\nDataset Info:")
df.info()
```

```python
print("\nStatistical Summary:")
df.describe()
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
print("Duplicate Rows:" , df.duplicated().sum())
ff = df.drop_duplicates()
```

```python
df.columns
plt.figure()
sns.histplot(df['Calories (kcal)'] , kde=True)
plt.title("Distribution Of Calories")
plt.show()
```

```python
df.columns
plt.figure()
sns.histplot(df['Protein (g)'] , kde=True)
plt.title("Distribution Of Proteins")
plt.show()
```

```python
top_cal = df.sort_values(by='Calories (kcal)' , ascending=False).head(10)

plt.figure()
sns.barplot(x='Calories (kcal)' , y='Dish Name' , data=top_cal)
plt.title('Top 10 Dishes with Highest Calories')
plt.show()
```

```python
top_cal = df.sort_values(by='Carbohydrates (g)' , ascending=False).head(10)

plt.figure()
sns.barplot(x='Carbohydrates (g)' , y='Dish Name' , data=top_cal)
plt.title('Top 10 Dishes with Highest Carbohydrates')
plt.show()
```
