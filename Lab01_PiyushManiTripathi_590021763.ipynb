```python
import numpy as np
import pandas as pd
```

```python
dataset={"Students_name":["A","B","C","D","E","F"],
         "Marks":[65,54,np.nan,66,np.nan,78],
         "Attendance":[90,np.nan,85,120,np.nan,876]}
```

```python
df=pd.DataFrame(dataset)
```

```python
df
```

```python
df.isnull().sum()
```

```python
sum(df.isnull().sum())
```

Mean Imputation


```python
df["Marks"].fillna(df["Marks"].mean())
```

```python
df["Attendance"].fillna(df["Attendance"].mean())
```

Outlayer detection and handling

```python
df.loc[df["Attendance"]>100,"Attendance"]=100
```

Feature Scaling


```python
dataset={"Students_name":["A","B","C","D","E"],
         "Marks":[50,60,70,80,90]}
```

```python
df=pd.DataFrame(dataset)
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
df["Marks"]=scalar.fit_transform(df[["Marks"]])
```

```python
df
```

```python
df["shifted"]=df["Marks"]+10
```

```python
df
```

```python
df["logMarks"]=np.log(df["shifted"])
```
